# Q3.Using line chart Visualize the unemployment rate of workers by industry.

```python
for column in df.columns:
    if (column == 'datestamp'):
        continue
    plt.plot(df[column])
    plt.title(column + " Unemployment Rate")
    plt.ylabel("Unemployment rate")
    plt.xlabel("Year")
    plt.show()
```

# Q4.Plot the monthly and yearly trends .

```python
#monthly trend 

df['month'] = df.index.month
df['year'] = df.index.year

df_month = df.groupby(by = 'month').mean()

for column in df_month.columns:
    if ( column == 'month' or column == 'year'):
        continue
    sns.barplot(x = df_month.index, y = df_month[column])
    sns.lineplot(x = df_month.index, y = df_month[column])
    plt.show()
```

```python
#monthly trend 
months = df.resample("M").mean()
months.loc[:, months.columns != 'datestamp'].plot(subplots=True, figsize=(12, 15))
```

```python
#yearly trend

df_year = df.groupby(by = 'year').mean()

print(df_year.index)

for column in df_year.columns:
    if ( column == 'month' or column == 'year'):
        continue
    sns.barplot(x = df_year.index, y = df_year[column])
    plt.show()
```

```python
#yearly trend
years = df.resample("Y").mean()
years.loc[:, years.columns != 'datestamp'].plot(subplots=True, figsize=(12, 15))
```

# Q5.Apply time series decomposition to your dataset to visualize the trend and seasonality.

```python
import matplotlib.dates as mdates
import statsmodels.api as sm

df["total"] = df["Agriculture"] + df["Business services"] + df["Construction"] + df["Durable goods manufacturing"] + df["Education and Health"] + df["Finance"] + df["Government"] + df["Information"] + df["Leisure and hospitality"] + df["Manufacturing"] + df["Mining and Extraction"] + df["Nondurable goods manufacturing"] + df["Other"] + df["Self-employed"] + df["Transportation and Utilities"] + df["Wholesale and Retail Trade"]

decomposition = sm.tsa.seasonal_decompose(df["total"].dropna())
decomposition.seasonal["2000":"2010"].plot()
```

```python
pd.DataFrame(decomposition.trend).plot(subplots=True, layout=(2, 1), linewidth=3);
```

# Q6. Visualize the seasonality of Agriculture, Health and Finance sector.

```python
from  statsmodels.tsa.seasonal import seasonal_decompose
sep_clmn = ["Agriculture", "Education and Health", "Finance"]
i = 0
for (columnName, columnData) in df[sep_clmn].iteritems():
    f = plt.figure()
    a = f.add_subplot(111)
    a.set_title(df[sep_clmn].columns[i])
    res = seasonal_decompose(df[columnName], model='additive')
    res.seasonal.plot()
    plt.legend(loc="upper right")
    i += 1
```

# Q7. Visualize the seasonality of multiple time series and the correlation between each time series in the dataset.

```python
results = []
i = 0
for (columnName, columnData) in df.iteritems():
    f = plt.figure()
    a = f.add_subplot(111)
    res = seasonal_decompose(df[columnName], model='additive')
    a.set_title(df.columns[i])
    res.seasonal.plot()
    plt.legend(loc="upper right")
    results.append(res.seasonal)
    i += 1
```

```python
results = pd.DataFrame(results)
results.corr()
```
